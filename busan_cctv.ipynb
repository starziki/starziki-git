```
pop_df = pd.read_csv('201905_201905_연령별인구현황_월간.csv',encoding='euc-kr')
cctv_df = pd.read_csv('부산광역시_CCTV_20190128.csv',encoding='euc-kr',)

pop_df.drop([0], axis=0, inplace=True) #부산광역시 미리 총계 낸 값 삭제
pop_df
```

```
pop_df['행정구'] = pop_df['행정구역'].str.split(' ', n=2, expand=True)[1] #구로 추출
pop_df=pop_df.set_index('행정구') #인덱스를 잡음
pop_df
```

```
cctv_df['소새지구']=cctv_df['소재지도로명주소'].str.split(' ',n=2,expand=True)[1] # 구로 추출
cctv_count=cctv_df['소새지구'].groupby(cctv_df['소새지구']).count() #구로 묶어서 개수를 셈
```

```
pop_df['cctv']=cctv_count #해당 index에 맞는 cctv의 개수를 위에 잡아놓은 인덱스로 넣음
```
